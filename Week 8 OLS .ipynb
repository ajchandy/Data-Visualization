```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
Location = "C:/Akhil/USF BAIS/Data Vis/gradedata.xlsx"
df = pd.read_excel(Location)
df.tail()
```

```python
df1=df[['grade','exercise', 'hours', 'gender']]
df1= pd.get_dummies(df1, drop_first=True)
result = sm.ols(formula='grade ~ exercise + hours + gender_male',data=df1).fit()
result.summary()
```
