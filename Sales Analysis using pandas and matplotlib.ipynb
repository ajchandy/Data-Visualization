```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''Display a table of sales data using pandas '''
salesMen = ['Ahmed', 'Omar', 'Ali', 'Ziad', 'Salwa', 'Lila']
Mobile_Sales = [2540, 1370, 1320, 2000, 2100, 2150]
TV_Sales = [2200, 1900, 2150, 1850, 1770, 2000]
df = pd.DataFrame()
df ['Name'] =salesMen
df ['Mobile_Sales'] = Mobile_Sales
df['TV_Sales']=TV_Sales
df.set_index("Name",drop=True,inplace=True)
df
```

```python
'''Graphs using Matplotlib'''
'''Sales by volume for each salesman bar'''
df.plot.bar( figsize=(20, 10), rot=0).legend(bbox_to_anchor=(1.1, 1))
plt.xlabel('Salesmen') 
plt.ylabel('Sales')
plt.title('Sales Volume for two salesmen in \nJanuary and April 2017')
plt.show()
```

```python
'''Pie Chart of item sales'''
df.plot.pie(subplots=True)
```

```python
'''Box plot of item sales'''
df.plot.box()
plt.xlabel('Items') 
plt.ylabel('Number of units')
plt.title('Box plot of total sales by item')
```

```python
'''Area plot of item sales'''
df.plot.area(figsize=(6, 4)).legend(bbox_to_anchor=(1.3,1))
plt.xlabel('Salesman') 
plt.ylabel('Sales')
plt.title('Stacked bar plot sales by salesman')
```

```python
'''Stacked Bar plot'''
df.plot.bar(stacked=True, figsize=(20, 10)).legend\
(bbox_to_anchor=(1.1, 1))
plt.xlabel('Salesman') 
plt.ylabel('Sales')
plt.title('Stacked bar plot sales by salesman')
```
