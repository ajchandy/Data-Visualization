```python
import pandas as pd
import numpy as np
Location = "C:/Users/akhil/Downloads/gradedata.xlsx"
df = pd.read_excel(Location)
```

```python
#Create yes or no based on conditions
df['isFailing'] = np.where(df['grade']<70,
'yes', 'no')
df.tail(5)
```

```python
#Create Yes/No isFailingMale Column
df['isFailingMale'] = np.where((df['grade']<70)
& (df['gender'] == 'male'),'yes', 'no')
df.tail(5)
```

```python
#Create time management column
df['fortimemgmt'] = np.where((df['exercise']>3)
& (df['hours'] > 17),'busy', 'not_busy')
df.tail(5)
```
